```python
import pandas as pd
import joblib
from sklearn.preprocessing import MinMaxScaler

df = pd.read_parquet("../data/interim/01_loaded.parquet")

print(df.shape)
df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])

df = df.sort_values("timestamp").reset_index(drop=True)

df.head()
```

```python
X = df[["temperature", "humidity", "co2"]]

y = df["yield_kg"]

print("X Shape:", X.shape)
print("y Shape:", y.shape)
```

```python
split_idx = int(len(df) * 0.8)

X_train = X.iloc[:split_idx]
X_test = X.iloc[split_idx:]

y_train = y.iloc[:split_idx]
y_test = y.iloc[split_idx:]
```

```python
print("Train Rows:", len(X_train))
print("Test Rows :", len(X_test))

print("\nTrain Period:")
print(df.iloc[0]["timestamp"], "to",
      df.iloc[split_idx-1]["timestamp"])

print("\nTest Period:")
print(df.iloc[split_idx]["timestamp"], "to",
      df.iloc[-1]["timestamp"])
```

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Scaling completed")
```

```python
joblib.dump(scaler, "../models/scaler.joblib")

print("Scaler saved successfully")
```

```python
pd.DataFrame(
    X_train_scaled,
    columns=X.columns
).to_csv("../data/processed/X_train.csv", index=False)

pd.DataFrame(
    X_test_scaled,
    columns=X.columns
).to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)

y_test.to_csv("../data/processed/y_test.csv", index=False)

print("All files saved successfully")
```
